```python
import numpy as np
import pandas as pd
df = pd.read_csv('numberseq.csv', dtype=object)
df
```

```python
del df['scale']
del df['sequence']
```

```python
df['scale'] = ''
df['mag'] = pd.to_numeric(df['mag'])
df['scale'][df['mag']<4] = '0'
df['scale'][(df['mag']>=4) & (df['mag']<4.5)] = '1'
df['scale'][df['mag']>=4.5] = '2'
```

```python
df['phase'] = pd.to_numeric(df['phase'])
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    if lunation in range(0, 7):
        return 0
    elif lunation in range (7, 15):
        return 1
    elif lunation in range (15, 21):
        return 2
    elif lunation in range (21, 30):
        return 3
    #return lunation

df['startdate'] = pd.to_datetime(df.startdate)
df['startdate'] = df['startdate'].dt.date
for i in range(0, len(df)):
    df['phase'][i] = get_phase_on_day(df['startdate'][i])
```

```python
df
```

```python
df['dep'] = ''
df['dep'][df['depth']<70] = '0'
df['dep'][(df['depth']>= 70)&(df['depth']<300)] = '1'
df['dep'][df['depth']>=300] = '2'
df['sequence'] = ''
df['sequence'] = df['type'] + df['cluster'] + df['scale'] + df['dep']
df
```

```python
df['Latitude'] = pd.to_numeric(df['Latitude'])
df['Longitude'] = pd.to_numeric(df['Longitude'])
df.round({'Latitude': 3, 'Longitude': 3})
df = df.sort_values(by = ['startdate', 'sequence'])
df.to_csv('numberseq.csv', index = False)
```

```python
with open("sequence.txt", "a") as myfile:
    myfile.write(df['sequence'][0])
for x in range(1, len(df)):
    #if date is different from the upcoming one
    if (df['startdate'][x] == df['startdate'][x - 1] and df['sequence'][x] == df['sequence'][x - 1]):
        continue
    elif ((df['startdate'][x] == df['startdate'][x - 1]) and (df['sequence'][x] != df['sequence'][x - 1])):
        with open("sequence.txt", "a") as myfile:
            myfile.write(' ' + df['sequence'][x])
    elif (df['startdate'][x] != df['startdate'][x - 1]):
        with open("sequence.txt", "a") as myfile:
            myfile.write(' ' + str(df['phase'][x-1]) + ' -1 ')
            #' ' + str(result['phase'][x-1]) +
        if (df['phase'][x - 1] != df['phase'][x]):
            with open("sequence.txt", "a") as myfile:
                myfile.write('-2\n')
        with open("sequence.txt", "a") as myfile:
            myfile.write(df['sequence'][x])
        
with open("sequence.txt", "a") as myfile:
        myfile.write(' -1 -2\n')
```
