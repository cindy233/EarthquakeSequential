```python
import numpy as np
import pandas as pd
file = open('catsearch1.csv', 'r')
mags = []
datetimes = []
years = []
months = []
days = []
clus = []
types = []

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')
    datetime = str(data[0])
    datetimes.append(datetime)
    year = str(data[0])[:4]
    years.append(year)
    month = str(data[0])[5:7]
    months.append(month)
    day = str(data[0])[8:10]
    days.append(day)
    mag = float(data[1])
    mags.append(mag)
    clu = int(data[4])
    clus.append(clu)
    typ = str('E')
    types.append(typ)
    
mags = np.asarray(mags)
datetimes = np.asarray(datetimes)
years = np.asarray(years)
months = np.asarray(months)
days = np.asarray(days)
clus = np.asarray(clus)
types = np.asarray(types)
```

```python
all_data = {'startyear': years, 'startmonth': months, 'startday': days, 'mag': mags, 'cluster': clus, 'type': types}
```

```python
df = pd.DataFrame(all_data)
df
```

```python
df2 = df.query('startyear == "2001" and mag >= 6')
```
