```python
import numpy as np
import pandas as pd
from geopy.distance import vincenty

file = open('centroids.csv', 'r')
centroids =[]

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')
    centroid = (float (data[0]), float(data[1]))
    centroids.append(centroid)
    
centroids = np.asarray(centroids)
```

```python

file = open('GVP_Eruption_Results.csv', 'r')

points = []
numbers = []
names = []
styears = []
stmonths = []
stdays = []
edyears = []
edmonths = []
eddays = []
veis = []
latitudes = []
longitudes = []

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')    
    point = (float(data[-2]), float(data[-1]))
    points.append(point)
    latitude = float(data[-2])
    latitudes.append(latitude)
    longitude = float(data[-1])
    longitudes.append(longitude)
    number = int(data[0])
    numbers.append(number)
    name = str(data[1])
    names.append(name)
    vei = str(data[5])
    veis.append(vei)
    styear = str(data[8])
    styears.append(styear)
    stmonth = str(data[10])
    stmonths.append(stmonth)
    stday = str(data[12])
    stdays.append(stday)
    

points = np.asarray(points)
numbers = np.asarray(numbers)
names = np.asarray(names)
veis = np.asarray(veis)
styars = np.asarray(styears)
stmonths = np.asarray(stmonths)
stdays = np.asarray(stdays)
latitudes = np.asarray(latitudes)
longitudes = np.asarray(longitudes)
```

```python
ind = []
for point in points:
    dista = []
    for centroid in centroids:
        dista.append(vincenty(point, centroid).miles)
    ind.append(dista.index(min(dista)))

inds = np.asarray(ind)
```

```python
all_data = {'startyear': styears, 'startmonth': stmonths, 'startday': stdays, 'VEI': veis, 'latitude': latitudes, 'longitude': longitudes, 'cluster': inds}
df = pd.DataFrame(all_data)
df
```

```python
df.to_csv('volcluster.csv')
```
