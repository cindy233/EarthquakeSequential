```python
import numpy as np
import pandas as pd
df1 = pd.read_csv('basket.csv')
df1
```

```python
df1['sequence']=df1['type'] + df1['cluster'].map(str)
```

```python
with open("test.txt", "a") as myfile:
        myfile.write(df1['sequence'][0])
for x in range(1, len(df1['sequence'])):
    #if date is different from the upcoming one
    if ((df1['startdate'][x] != df1['startdate'][x - 1]) and (df1['sequence'][x] != df1['sequence'][x - 1])):
        with open("test.txt", "a") as myfile:
            myfile.write(' -1 ' + df1['sequence'][x] )
    elif df1['sequence'][x] != df1['sequence'][x - 1]:
        with open("test.txt", "a") as myfile:
            myfile.write(' ' + df1['sequence'][x])
with open("test.txt", "a") as myfile:
        myfile.write(' -1 -2\n')
```
