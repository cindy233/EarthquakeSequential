```python
import numpy as np
import pandas as pd
df = pd.read_csv('wtemp.csv', dtype=object)
df
```

```python
dff = df.copy()
```

```python
dfsp1 = df[df['type'] == '1'].sample(800)
dfsp2 = df[df['type'] == '3'].sample(800)
frames = [dfsp1, df[df['type'] == '2'], dfsp2]
result = pd.concat(frames)
result['startdate'] = pd.to_datetime(result['startdate'])
result = result.sort(['startdate'])
result.index = range(len(result))
result
```

```python
#result = df.copy()
with open('samplesequence3.txt', "a") as myfile:
    myfile.write(result['sequence'][0])
    #str(result['phase'][0]) + ' ' + str(result['temp'][0]) + ' -1 ' + 
    for x in range(1, len(result)):
        #if date is the same from the previous one
        if ((result['startdate'][x] == result['startdate'][x - 1]) and (result['sequence'][x] == result['sequence'][x - 1])):
            continue
        elif ((result['startdate'][x] == result['startdate'][x - 1]) and (result['sequence'][x] != result['sequence'][x - 1])):
            myfile.write(' ' + result['sequence'][x])
        #if date is diff from the previous one
        elif (result['startdate'][x] != result['startdate'][x - 1]):
            if (result['phase'][x - 1] > result['phase'][x]):
                myfile.write(' -1 -2\n' + result['sequence'][x])
                #str(result['phase'][x]) + ' ' + str(result['temp'][x]) + ' -1 ' +
            else:
                myfile.write(' -1' + ' ' + str(result['sequence'][x]))
    myfile.write(' -1 -2\n')
```
