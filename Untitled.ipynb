```python
import numpy as np
import pandas as pd
df = pd.read_csv('wtemp.csv', dtype=object)
df
```

```python
dff = df.copy()
```

```python
df = dff[dff['type'] == '1']
result = df[(df['cluster']== '01') | (df['cluster']== '03') | (df['cluster']== '04') | (df['cluster']== '06') | (df['cluster']== '07') | (df['cluster']== '08')| (df['cluster']== '10')| (df['cluster']== '11')] 
result.index = range(len(result))
with open('tempsequence1.txt', "a") as myfile:
    myfile.write(str(result['phase'][0]) + ' ' + str(result['temp'][0]) + ' -1 ' + result['sequence'][0])
    for x in range(1, len(result['startdate'])):
        #if date is the same from the previous one
        if (result['startdate'][x] == result['startdate'][x - 1] and result['sequence'][x] == result['sequence'][x - 1]):
            continue
        elif ((result['startdate'][x] == result['startdate'][x - 1]) and (result['sequence'][x] != result['sequence'][x - 1])):
            myfile.write(' ' + result['sequence'][x])
        #if date is diff from the previous one
        elif (result['startdate'][x] != result['startdate'][x - 1]):
            if (result['phase'][x - 1] != result['phase'][x]):
                myfile.write(' -1 -2\n' + str(result['phase'][x]) + ' ' + str(result['temp'][x]) + ' -1 ' + result['sequence'][x])
            else:
                myfile.write(' -1' + ' ' + str(result['sequence'][x]))
    myfile.write(' -1 -2\n')
```

```python
df['temp'] = ''
df['pavg'] = pd.to_numeric(df['pavg'])
df['temp'][df['pavg'] < 0] = 4
df['temp'][(df['pavg']>=0) & (df['pavg']<0.5)] = 5
df['temp'][(df['pavg']>=0.5) & (df['pavg']<1)] = 6
df['temp'][df['pavg']>=1] = 7
df
```

```python
df['depth'] = pd.to_numeric(df['depth'])
df['dep'] = ''
df['dep'][df['depth']<70] = '0'
df['dep'][(df['depth']>= 70)&(df['depth']<300)] = '1'
df['dep'][df['depth']>=300] = '2'
df
```

```python
df = df.fillna('')
df['sequence'] = ''
df['sequence'] = df['type'] + df['cluster'] + df['scale'] + df['dep']
```
