```python
import numpy as np
import pandas as pd
df1 = pd.read_csv('allvolcanoes.csv', dtype=object)
df1['startdate'] = df1.apply(lambda row: '{0}/{1}/{2}'.format(row['Start_Month'],row['Start_Day'],row['Start_Year']), axis=1)
#print(df1['startdate'])
df1['startdate'] = pd.to_datetime(df1.startdate)
df1 = df1.sort_values(by = ['startdate'])
df1['type'] = '3'
df1
```

```python
df = pd.read_csv('allquakeswnphase.csv', dtype=object)
#df['startdate'] = df.apply(lambda row: '{0}/{1}/{2}'.format(row['startyear'],row['startmonth'],row['startday']), axis=1)
df['startdate'] = pd.to_datetime(df.startdate)
df['Longitude'] = pd.to_numeric(df['Longitude'])
df['Latitude'] = pd.to_numeric(df['Latitude'])
from datetime import timedelta
df['enddate'] = df['startdate'] + timedelta(days=30)
#result['startdate'] = result['startdate'].dt.date
df = df.sort_values(by = ['startdate'])
df.index = range(len(df))
df
```

```python
dff = df.copy()
```

```python
from geopy.distance import great_circle
for i in range(0, len(df)-1): 
    df2 = df[(df['startdate'] >= df['startdate'][i]) & (df['startdate'] < df['enddate'][i]) & (df['Latitude'] >= df['Latitude'][i] - 2) & (df['Latitude'] < df['Latitude'][i] + 2)]
    if len(df2) >= 30:
        counter = 0
        l = []
        for j in range (i + 1 , i + len(df2)):
            if great_circle((df['Latitude'][i], df['Longitude'][i]), (df['Latitude'][j], df['Longitude'][j])).miles < 100:
                counter = counter + 1
                l.append(j)
        if counter >= 30:
            #print i
            #print counter
            #return index entry of df2[0]
            y = list(df2.index.values)[0]
            df['type'][y] = '2'
            df = df.drop(df.index[l])
            df.index = range(len(df))
```

```python
df.to_csv('quakeswarm.csv',index = False)
df
```

```python
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    if lunation in range(4, 11):
        return 1
    elif lunation in range (11, 19):
        return 2
    elif lunation in range (19, 27):
        return 3
    else:
        return 0
    #return lunation

result['startdate'] = pd.to_datetime(result.startdate)
result['startdate'] = result['startdate'].dt.date
result['phase'] = ''
for i in range(0, len(result)):
    result['phase'][i] = get_phase_on_day(result['startdate'][i])
```
