```python
from datetime import datetime
import ephem
def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    return lunation
    #if (lunation >= 0 and lunation <7):
def nnm(date):
    d = ephem.Date(date)
    nnm = ephem.next_new_moon(date)
    return int(nnm - d)

temp['Anomaly'] = pd.to_numeric(temp['Anomaly'])
temp['pavg'] = ''
for i in range(0, len(temp) - 1):
    phase = get_phase_on_day(temp['date'][i])
    
    if (phase >= 26):
        end = nnm(temp['date'][i])
        temp['pavg'][i] = temp['Anomaly'].iloc[i - phase + 26: i + end + 4].mean(axis=0)
    elif (phase < 4):
        temp['pavg'][i] = temp['Anomaly'].iloc[i - phase - 4 : i - phase + 4].mean(axis=0)
        #print temp['pavg'][i]
    elif (phase >= 4 and phase < 11):
        temp['pavg'][i] = temp['Anomaly'].iloc[i - phase + 4: i - phase + 11].mean(axis=0)
    elif (phase >= 11 and phase < 19):
        temp['pavg'][i] = temp['Anomaly'].iloc[i - phase + 11: i - phase + 19].mean(axis=0)
    elif (phase >= 19 and phase < 26):
        temp['pavg'][i] = temp['Anomaly'].iloc[i - phase + 19: i - phase + 26].mean(axis=0)

temp['pavg'][0] = -1.193
temp['pavg'][42002] = 0.898
temp
```

```python
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    if lunation in range(0, 8):
        return 0
    elif lunation in range (8, 16):
        return 1
    elif lunation in range (16, 24):
        return 2
    elif lunation in range (24, 30):
        return 3
    #return lunation

result['startdate'] = pd.to_datetime(result.startdate)
result['startdate'] = result['startdate'].dt.date
for i in range(0, len(result)):
    if result['phase'][i] == '':
        result['phase'][i] = get_phase_on_day(result['startdate'][i])
```
