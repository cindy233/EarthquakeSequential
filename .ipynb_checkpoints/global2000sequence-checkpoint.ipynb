```python
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    return lunation

result['startdate'] = pd.to_datetime(result.startdate)
result['startdate'] = result['startdate'].dt.date
result['phase'] = ' '
for i in range(0, len(result)):
    result['phase'][i] = get_phase_on_day(result['startdate'][i])
```

```python
with open("test2000.txt", "a") as myfile:
        myfile.write(result['sequence'][0])
for x in range(1, len(result['sequence'])):
    #if date is different from the upcoming one
    if ((result['startdate'][x] != result['startdate'][x - 1]) and (result['sequence'][x] != result['sequence'][x - 1])):
        with open("test2000.txt", "a") as myfile:
            myfile.write(' ' + str(result['phase'][x-1]) + ' -1 ' + result['sequence'][x] )
    elif result['sequence'][x] != result['sequence'][x - 1]:
        with open("test2000.txt", "a") as myfile:
            myfile.write(' ' + result['sequence'][x])
with open("test2000.txt", "a") as myfile:
        myfile.write( ' ' + str(result['sequence'][len(result['phase']) - 1]) +' -1 -2\n')
```
