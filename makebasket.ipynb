```python
import numpy as np
import pandas as pd
import csv

df1 = pd.read_csv('2001volcanoes.csv')
df1['startdate'] = df1.apply(lambda row: '{0}/{1}/{2}'.format(row['startmonth'],row['startday'],row['startyear']), axis=1)
df1['startdate'] = pd.to_datetime(df1.startdate)
df1 = df1.sort_values(by = ['startdate'])
df1
```

```python
df1.to_csv('2001volcanoes.csv', index = False)
```

```python
df = pd.read_csv('2001magbigger6.csv')
df['startdate'] = df.apply(lambda row: '{0}/{1}/{2}'.format(row['startyear'],row['startmonth'],row['startday']), axis=1)
df['startdate'] = pd.to_datetime(df.startdate)
df = df.sort_values(by = ['startdate'])
df
```

```python
df.to_csv('2001magbigger6.csv', index = False)
```

```python
df = df.iloc[:,(0,-1,-3)]
df1 = df1.iloc[:,(1,-1,-3)]
```

```python
frames = [df1, df]
result = pd.concat(frames)
result
```
