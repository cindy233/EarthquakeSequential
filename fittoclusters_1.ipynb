```python
import numpy as np
import pandas as pd
from geopy.distance import vincenty

file = open('centroids.csv', 'r')
centroids =[]

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')
    centroid = (float (data[2]), float(data[1]))
    centroids.append(centroid)
    
centroids = np.asarray(centroids)
print(centroids)
```

```python
df1 = pd.read_csv('catsearch.csv', dtype = object)
df1
```

```python
file = open('catsearch.csv', 'r')

points = []
latitudes = []
longitudes = []

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')    
    point = (float(data[1]), float(data[2]))
    points.append(point)
    latitude = float(data[1])
    latitudes.append(latitude)
    longitude = float(data[2])
    longitudes.append(longitude)
    
points = np.asarray(points)
latitudes = np.asarray(latitudes)
longitudes = np.asarray(longitudes)
```

```python
from geopy.distance import great_circle
ind = []
for point in points:
    dista = []
    for centroid in centroids:
        dista.append(great_circle(point, centroid).miles)
    ind.append(dista.index(min(dista)))

inds = np.asarray(ind)
```

```python
df1['cluster'] = inds
df1
```

```python
df1.to_csv('catsearchwcluster.csv', index = False)
```
