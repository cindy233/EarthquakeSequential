```python
result['sequence']=result['type'] + result['cluster'].map(str)
result = result.sort_values(by = ['startdate', 'sequence'])
```

```python
result['seqdep'] = result['sequence']
for i in range (0,len(result['startdate'])):
    result['depth'][i] = float(result['depth'][i])
    if (result['depth'][i] >= 70 and result['depth'][i] < 300):
        result['seqdep'][i] = str(result['sequence'][i]) + "D"
    elif (result['depth'][i] >=300 ):
        result['seqdep'][i] = str(result['sequence'][i])+ "DD"
result
```

```python
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    if lunation in range(0, 8):
        return 0
    elif lunation in range (8, 16):
        return 1
    elif lunation in range (16, 24):
        return 2
    elif lunation in range (24, 30):
        return 3
    #return lunation

result['startdate'] = pd.to_datetime(result.startdate)
result['startdate'] = result['startdate'].dt.date
result['phase'] = ' '
for i in range(0, len(result)):
    result['phase'][i] = get_phase_on_day(result['startdate'][i])
```

```python
with open("test2001plus.txt", "a") as myfile:
        myfile.write(result['sequence'][0])
for x in range(1, len(result['sequence'])-1):
    #if date is different from the upcoming one
    if (result['startdate'][x] == result['startdate'][x - 1] and result['sequence'][x] == result['sequence'][x - 1]):
        continue
    elif ((result['startdate'][x] == result['startdate'][x - 1]) and (result['sequence'][x] != result['sequence'][x - 1])):
        with open("test2001plus.txt", "a") as myfile:
            myfile.write(' ' + result['seqdep'][x])
    elif (result['startdate'][x] != result['startdate'][x - 1]):
        with open("test2001plus.txt", "a") as myfile:
            myfile.write(' ' + str(result['phase'][x-1]) + ' -1 ')
        if (result['phase'][x - 1] > result['phase'][x]):
            with open("test2001plus.txt", "a") as myfile:
                myfile.write('-2\n')
        with open("test2001plus.txt", "a") as myfile:
            myfile.write(result['seqdep'][x])
        
with open("test2001plus.txt", "a") as myfile:
        myfile.write( ' ' + str(result['phase'][len(result['phase'])]) +' -1 -2\n')
```
