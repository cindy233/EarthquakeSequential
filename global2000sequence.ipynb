```python
df = pd.read_csv('allquakes.csv', dtype=object)
df['startdate'] = df.apply(lambda row: '{0}/{1}/{2}'.format(row['startyear'],row['startmonth'],row['startday']), axis=1)
df['startdate'] = pd.to_datetime(df.startdate)
#result['startdate'] = result['startdate'].dt.date
df = df.sort_values(by = ['startdate'])
df
```

```python
df_eq = df.iloc[:,(0,-1,-2,1,2,3,4)]
df_vo = df1.iloc[:,(-3,-1,-2,5,-5,-4)]
```

```python
frames = [df_eq, df_vo]
result = pd.concat(frames)
result
```

```python
result['sequence']=result['type'] + result['cluster'].map(str)
result = result.sort_values(by = ['startdate', 'sequence'])
result
```

```python
result.index = range(len(result))
result
```

```python
result['scale'] = ''

for i in range (0, len(result['sequence'])):
    result['mag'][i] = float(result['mag'][i])
    if (result['mag'][i] <= 3.0):
        result['scale'][i] = 'L'
    elif (result['mag'][i] > 3.0 and result['mag'][i] <= 6.0):
        result['scale'][i] = 'M'
    elif (result['mag'][i] > 6.0):
        result['scale'][i] = 'H'
result
#type(result['depth'][3])
```

```python
result['seqdep'] = result['sequence']
for i in range (0,len(result['startdate'])):
    result['depth'][i] = float(result['depth'][i])
    if (result['depth'][i] >= 70 and result['depth'][i] < 300):
        result['seqdep'][i] = 'I'
    elif (result['depth'][i] >=300 ):
        result['seqdep'][i] = 'D'
result
```

```python
result.to_csv('scaleseqdep.csv',index = False)
```

```python
from datetime import datetime
import ephem

def get_phase_on_day(date):
    d = ephem.Date(date)
    pnm = ephem.previous_new_moon(date)
    lunation = int(d - pnm)
    if lunation in range(0, 8):
        return 0
    elif lunation in range (8, 16):
        return 1
    elif lunation in range (16, 24):
        return 2
    elif lunation in range (24, 30):
        return 3
    #return lunation

result['startdate'] = pd.to_datetime(result.startdate)
result['startdate'] = result['startdate'].dt.date
result['phase'] = ' '
for i in range(0, len(result)):
    result['phase'][i] = get_phase_on_day(result['startdate'][i])
```
