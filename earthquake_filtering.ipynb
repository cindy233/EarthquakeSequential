```python
import numpy as np
import pandas as pd
file = open('catsearch.csv', 'r')
mags = []
datetimes = []
latitudes = []
longitudes = []
years = []
months = []
days = []

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')
    datetime = str(data[0])
    datetimes.append(datetime)
    year = str(data[0])[:4]
    years.append(year)
    month = str(data[0])[5:7]
    months.append(month)
    day = str(data[0])[8:10]
    days.append(day)
    latitude = float (data[1])
    latitudes.append(latitude)
    longitude = float (data[2])
    longitudes.append(longitude)
    mag = float(data[4])
    mags.append(mag)

mags = np.asarray(mags)
datetimes = np.asarray(datetimes)
latitudes = np.asarray(latitudes)
longitudes = np.asarray(longitudes)
years = np.asarray(years)
months = np.asarray(months)
days = np.asarray(days)
```

```python
all_data = {'year': years, 'month': months, 'day': days, 'mag': mags, 'latitude': latitudes, 'longitude': longitudes}
```

```python
df = pd.DataFrame(all_data)
df
```

```python
df2 = df.query('year == "2001" or year == "2002" or year == "2003" or year == "2004" or year == "2005"')
df2
```

```python
df2.to_csv('filter.csv')
```
