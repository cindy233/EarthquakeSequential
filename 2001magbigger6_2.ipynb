```python
import numpy as np
import pandas as pd
file = open('catsearchwcluster.csv', 'r')
mags = []
datetimes = []
years = []
months = []
days = []
clus = []
types = []
depths = []
lats = []
longs = []

firstline = True
for line in file:
    if firstline:    #skip first line
        firstline = False
        continue
    data = line.strip().split(',')
    datetime = str(data[0])
    datetimes.append(datetime)
    year = str(data[0])[:4]
    years.append(year)
    month = str(data[0])[5:7]
    months.append(month)
    day = str(data[0])[8:10]
    days.append(day)
    mag = float(data[4])
    mags.append(mag)
    depth = str(data[3])
    depths.append(depth)
    clu = int(data[-1])
    clus.append(clu)
    typ = str('1')
    types.append(typ)
    latitude = float(data[1])
    lats.append(latitude)
    longitude = float(data[2])
    longs.append(longitude)
    
mags = np.asarray(mags)
depths = np.asarray(depths)
datetimes = np.asarray(datetimes)
years = np.asarray(years)
months = np.asarray(months)
days = np.asarray(days)
clus = np.asarray(clus)
types = np.asarray(types)
lats = np.asarray(lats)
longs = np.asarray(longs)
```

```python
all_data = {'startyear': years, 'startmonth': months, 'startday': days, 'depth': depths, 'mag': mags, 'cluster': clus, 'type': types, 'latitude': lats, 'longitude': longs}
```

```python
df = pd.DataFrame(all_data)
df.to_csv('earthquakesimplified.csv', index = False)
```
