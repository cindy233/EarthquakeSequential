```python
import numpy as np
import pandas as pd
import csv

df1 = pd.read_csv('2001volcanoes.csv', dtype=object)
df1['startdate'] = df1.apply(lambda row: '{0}/{1}/{2}'.format(row['Start_Month'],row['Start_Day'],row['Start_Year']), axis=1)
print(df1['startdate'])
df1['startdate'] = pd.to_datetime(df1.startdate)
df1 = df1.sort_values(by = ['startdate'])
df1['type'] = 'V'
df1
```

```python
df1.to_csv('2001volcanoes.csv', index = False)
```
